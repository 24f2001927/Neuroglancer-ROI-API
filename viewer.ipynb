```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

def view_extracted_roi(file_path):
    """
    Loads a 3D NumPy array from disk and creates an interactive 
    slice viewer in the notebook.
    """
    try:
        # Load the saved ROI
        volume = np.load(file_path)
    except FileNotFoundError:
        print(f"[ERROR] Could not find the file: {file_path}")
        return
    except Exception as e:
        print(f"[ERROR] Failed to load file: {e}")
        return

    # Verify it's a 3D volume
    if volume.ndim != 3:
        print(f"[ERROR] Expected a 3D array, but got a {volume.ndim}D array with shape {volume.shape}.")
        return

    print(f"[SUCCESS] Loaded ROI with shape: {volume.shape} (Z, Y, X)")

    # The plotting function tied to the slider
    def plot_slice(z_index):
        fig, ax = plt.subplots(figsize=(8, 8))
        # Display the specific Z-slice using a grayscale colormap
        img = ax.imshow(volume[z_index, :, :], cmap='gray')
        ax.set_title(f"Z-Slice: {z_index} / {volume.shape[0] - 1}")
        ax.axis('off')
        
        # Optional: Add a colorbar to understand pixel intensity values
        plt.colorbar(img, ax=ax, fraction=0.046, pad=0.04)
        plt.show()

    # Create the interactive slider
    z_slider = widgets.IntSlider(
        min=0, 
        max=volume.shape[0] - 1, 
        step=1, 
        value=volume.shape[0] // 2, # Start in the middle of the volume
        description='Z-Slice:',
        layout=widgets.Layout(width='500px')
    )

    # Bind the slider to the plotting function
    widgets.interact(plot_slice, z_index=z_slider)
```

```python
# Replace with the actual file path generated by your FastAPI app
roi_path = "/path/to/roi_X_Y_Z.npy"

view_extracted_roi(roi_path)
```
